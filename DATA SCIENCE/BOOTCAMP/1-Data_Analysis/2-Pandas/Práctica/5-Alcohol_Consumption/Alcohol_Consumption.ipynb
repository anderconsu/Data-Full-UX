### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby('continent').mean()[['beer_servings']].sort_values(by = 'beer_servings', ascending= False)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent').describe()['wine_servings'].sort_values(by = 'mean', ascending= False)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continent').mean().sort_values(by= 'total_litres_of_pure_alcohol', ascending= False)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby('continent').median().sort_values(by= 'total_litres_of_pure_alcohol', ascending= False)
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
drinks.groupby('continent').aggregate(['mean', 'min', 'max'])['spirit_servings'].sort_values(by = 'mean', ascending= False)
```
