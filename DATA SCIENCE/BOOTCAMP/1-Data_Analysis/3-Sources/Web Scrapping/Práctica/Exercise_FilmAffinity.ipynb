```python
from bs4 import BeautifulSoup as bs
import requests
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

url = "https://www.filmaffinity.com/es/ranking.php?rn=ranking_fa_movies"

driver.get(url)
```

```python
Cookies = driver.find_element(by=By.XPATH,value='/html/body/div[1]/div/div/div/div[2]/div/button[2]')
Cookies.click()
#Cookies = driver.find_element("xpath", "/html/body/div[1]/div/div/div/div[2]/div/button[2]/span")
```

```python
boton = driver.find_element('xpath','/html/body/div[3]/div/div/main/div[2]/div[3]/i[1]')
```

```python
for i in range(0,8):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    boton.click()
    time.sleep(2)
```

```python
response = requests.get(url)
html = response.content
soup = bs(driver.page_source, "lxml")
soup
```

```python
rank = soup.select('li.position')
ranking = [int(r.get_text()) for r in rank]
ranking
```

```python
title = soup.select('div.mc-title')
name = [str(r.get_text()).split('(')[0] for r in title]
year = [str(r.get_text()).split('(')[1] for r in title]
name
```

```python
years = [(y.split(')')[0]) for y in year]
years
```

```python
pais = soup.select('div.mc-title img.nflag')
country = []
for a in pais:
       country.append(a.attrs.get('alt'))
country
```

```python
directores = soup.select('div.mc-director div.credits span.nb a')
director = [y.get_text() for y in directores]
director
```

```python
director.remove('Kátia Lund')
director.remove('Clyde Bruckman')
director.remove('Gene Kelly')
director.remove('Eric Toledano')
director.remove('Adrián Molina')
director.remove('Roger Allers')
director.remove('Lilly Wachowski')
director.remove('Lana Wachowski')
director.remove('Bob Peterson')
director.remove('Ingmar Bergman')
director.remove('Stuart Rosenberg')
```

```python
act = soup.select('div.mc-cast')
cast = [y.get_text() for y in act]
cast
```

```python
nota = soup.select('div.avg-rating')
rating = [float(y.get_text().replace(',','.')) for y in nota]
rating
```

```python
imagen = soup.select('div.mc-poster a')
poster = []
for a in imagen:
       poster.append(a.attrs.get('href'))
poster
```

```python
data = pd.DataFrame({'ranking':ranking,'name':name,'year':years,'country':country,'director':director,'cast':cast,'rating':rating,'poster':poster})
data
```

```python
data.to_csv('FilmAff',index = False)
```
